### 1 - Um laboratório de um hospital recebeu uma série de pedido de exames para descobrir se o sangue coletado possui um novo tipo de fungo no qual será reaizado um exame de sangue por dia para acompanhar uma suspeita de infecção fungica. Durante 7 dias será coletado sempre no mesmo horário uma pequena amostra de sangue e verificar, através de sorologia, o nível dos anti-corpos presentes no sangue. 
###A hipótese é, se a cada passo de tempo seguir aumentando então o paciente está infectado com esse fungo. Esse fungo só pode ser tratado depois de 7 dias da infecção **(JÁ RESPONDIDA)**

|Tempo de reação (dias)|Concentração anti-corpos(mg/L)|
|--|--|
|1|250|
|2|350|
|3|400|
|4|432|
|5|390|
|6|421|
|7|500|
>Obs: Para esse tipo de exame o valor normal varia entre 70-300 mg/L

```python
# Importa a biblioteca Numpy
import numpy as np 
# Importa o módulo para a geração do gráfico
from matplotlib import pyplot as plt 

# Definição de x e y 
x = np.arange(1, 8)
y = np.array([250, 350, 400, 432, 390, 421, 500])

# Gera a equação para a linha linear (Regressão linear)
equação = np.polyfit(x, y, 1)

# Captura a correlação entre os dois valores
correlação = np.corrcoef(x,y)[0][1]
# Cria a linha de tendência
linha_tendencia = np.poly1d(equação)

#---- Configurar gráfico ----

# Monta o texto para apresenta os resultados dentro do gráfico
texto = f'Linha de tendência: y = {equação[0]:.4f}x {equação[1]:.4f}'
texto += f'\nPearson (p): {correlação}'

# Define o tamanho da figura em polegadas
plt.figure(figsize=(6,5)) 

plt.title("Tempo em dias x Concentração de anti-corpos")
plt.xlabel("Tempo em dias")
plt.ylabel("Concentração de anti-corpos (mg/L)")

plt.text(0.05, 0.9, texto, 
         transform=plt.gca().transAxes)
plt.plot(x, y, 'ob')
plt.plot(x, linha_tendencia(x), 'r--')

plt.show()
```

###Exemplo de conclusão
Com base nos resultados apresentados no gráfico através de  ajuste de curvas e correlação de Pearson é possível aceitar a hipótese que a pessoa esteja infectada com o fungo uma vez que o coeficiente apresentou uma correlação forte (≈0.88) entre os dias passados e concentração de anti-corpos na corrente sanguinea. Isso sugere que há uma tendência de aumento de anti-corpos nos próximos dias caso não haja intervenção medicinal.

Obs: Não foram encontrados motivos para invalidar que há causalidade entre os dias passados e a concentração de anti-corpos. Os resultados, junto com análise do cenário manteve causalidade entre as variáveis até que seja feita uma contra prova.
