```python
# Carrega a biblioteca pandas e atribui um alias chamado "pd"
import pandas as pd
```

```python
# Deixar colunas e linhas aparentes
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Faz a leitura do arquivo train.csv
df = pd.read_csv('/content/drive/MyDrive/atividade/train.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df_menor = df[[
 'Id',
 'GarageCars',
 'GarageFinish',
 'GarageType',
 'GarageYrBlt',
 'GrLivArea',
 'LotArea',
 'LotFrontage',
 'MSSubClass',
 'MSZoning',
 'MasVnrArea',
 'SaleCondition',
 'SalePrice',
 'SaleType',
 'YrSold'
]]
df_menor.shape
```

```python
#2° forma de criar subplots
f, ax = plt.subplots(1,2, figsize=(10,5), dpi= 150)

ax[0].hist(df_menor['SalePrice'].dropna())
ax[0].set_title('Distribuição dos valores das casas')

ax[0].hist(df_menor['GarageType'].dropna())
ax[0].set_title('Distruibuiçao das garagens')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10), dpi=50)

plt.subplot(1,2,1) # 1 linha, 2 colucas, 1° Grafico
plt.hist(df_menor['SalePrice'].dropna()) #Histograma da variável Saleprice
plt.title('distribuicao dos valores das casa') # Titulo do Grafico 1


plt.subplot(1,2,1) # 1 linha, 2 colucas, 2° Grafico
plt.hist(df_menor['GarageType'].dropna()) #Histograma da variável Saleprice
plt.title('distribuicao das garagens') # Titulo do Grafico 1

plt.tight_layout()
plt.show
```

```python
# Seaborn
import seaborn as sns
```

```python

df_menor_2 = df_menor[[
 'Id',
 'GarageCars',
 'GarageFinish',
 'GarageType',
 'GarageYrBlt',
 'GrLivArea',
 'LotArea',
 'LotFrontage',
 'MSSubClass',
 'MSZoning',
 'MasVnrArea',
 'SaleCondition',
 'SalePrice',
 'SaleType',
 'YrSold'
]]
```

```python
# Gera distribuição de todas as variáveis
sns.pairplot(df_menor_2)
```

```python
# Boxplot (quartis) da variavel garageType e SalePrice com segmentação pela vriável MSZ.
plt.figure(figsize=(7,3), dpi=150)
sns.boxplot(x='MSZoning', y='SalePrice', hue='GarageFinish', data=df_menor)
```

```python
# Cria um objeto que terá as classes da variável passada, neste caso 5 classes (MSZoning)

g = sns.FacetGrid(df_menor, col='MSZoning')

# Gera um histograma SalePrice
g.map(plt.hist,'SalePrice', bins=10)
```
